```python
# imports
import pandas as pd
from pathlib import Path
```

```python
#load data
file_path =  Path('Resources/iris.csv')
iris_df = pd.read_csv(file_path)
iris_df.head()
```

```python
# drop a column
new_iris_df = iris_df.drop(['class'], axis=1)
new_iris_df.head()
```

```python
# convert columns to list in order to switch column order
titles = list(new_iris_df.columns)
titles
```

```python
# index the columns and switch positions
titles[1],titles[2] = titles[2], titles[1]
titles
```

```python
# place columns back into df
new_iris_df = new_iris_df[titles]
new_iris_df
```

```python
# Save file to csv
output_file_path = "Resources/new_iris_data.csv"
new_iris_df.to_csv(output_file_path, index=False)
```

```python
#load data
file_path =  Path('Resources/shopping_data.csv')
shopping_data_df = pd.read_csv(file_path)
shopping_data_df.head()
```

```python
# Columns
shopping_data_df.columns
```

```python
shopping_data_df.dtypes
```

```python
# find nullvalues
for column in shopping_data_df.columns:
    print(f"Column {column} has{shopping_data_df[column].isnull().sum()}\
null values")
```

```python
# Drop null rows
shopping_data_df = shopping_data_df.dropna()
```

```python
# Find duplicate entries
print(f"Duplicate entries: {shopping_data_df.duplicated().sum()}")
```

```python
# Remove the Customer ID Column
shopping_data_df.drop(columns= ["CustomerID"], inplace=True)
shopping_data_df.head()
```

```python
# Transform String column
def change_string(member):
    if member == "Yes":
        return 1
    else:
        return 0
shopping_data_df["Card Member"] = shopping_data_df["Card Member"].apply(change_string)
shopping_data_df
```

```python
# Transform annual income
shopping_data_df["Annual Income"] = shopping_data_df["Annual Income"] / 1000
shopping_data_df.head()
```

```python
shopping_data_df.rename(columns={"Card Member": "Card_Member", "Annual Income": "Annual_Income",
                                   "Spending Score (1-100)": "Spending_Score" }, inplace=True)
```
